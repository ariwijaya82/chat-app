```python
import numpy as np
from scipy.spatial import Voronoi, voronoi_plot_2d, distance_matrix
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import random
```

```python
fig, ax = plt.subplots()
rect = patches.Rectangle((0,0), 9, 6, linewidth=1, edgecolor='black', facecolor='green')
ax.add_patch(rect)

num_points = 5
random.seed(123)
x = [random.uniform(0,9) for _ in range(num_points)]
y = [random.uniform(0,6) for _ in range(num_points)]

x_ = [0,0,0,2.25,4.5,6.75,9,9,9,6.75,4.5,2.25]
y_ = [0,3,6,6,6,6,6,3,0,0,0,0]

plt.scatter(x, y, color='red', marker='o')
plt.scatter(x_, y_, color='pink', marker='o')

start_point = [0.5,0.5]
goal_point = [8.5,5.5]
plt.scatter(
  [start_point[0], goal_point[0]],
  [start_point[1], goal_point[1]],
  color='black',
  marker='o'
)
# voronoi diagram
vor = Voronoi(np.array([
    [start_point[0]]+x+x_+[goal_point[0]],
    [start_point[1]]+y+y_+[goal_point[1]]]).T)
voronoi_plot_2d(vor, ax=ax, show_points=False, line_colors='blue', line_alpha=0.5)

ax.set_aspect('equal')
ax.set_xlim(0, 9)
ax.set_ylim(0, 6)

plt.show()
```
